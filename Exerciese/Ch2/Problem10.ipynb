##### (c)

```r
library(repr)
options(repr.plot.width=8, repr.plot.height=5)
```

```r
hist(Boston$crim, breaks = 50)
```

So most suburban areas do not have crimes. Let's exclude them.

```r
Boston_red = Boston[Boston$crim < 20, ]
```

```r
par(mfrow=c(3, 5))

for (i in names(Boston_red)){
    plot(Boston_red[[i]], Boston_red$crim, xlab=i, ylab='crime')
}
```

There might be negative relationship between crime rate and chas, nox, dis, medv; and positive relationship between crime rate and indus, age, rad, tax, lstat.
